### PCA and PLS

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# import dataset
myDF = pd.read_excel("/Users/rauns/Downloads/lowarp.xlsx") # input your own dataset here, could be csv or excel
```

```python
myDF = myDF.fillna(myDF.mean())
myDF
```

```python
pca_df = myDF
pca_df.drop(["Primary ID", "Obs. Sec. ID:1"], axis=1, inplace=True)
pca_df
```

```python
scaler = StandardScaler()
scaler.fit(pca_df)
```

```python
scaled_data = scaler.transform(pca_df)
scaled_data
```

```python
pca = PCA(n_components = 4) # n components is the number of components you want to reduce the dimensions to 
```

```python
pca.fit(scaled_data)
```

```python
components = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
components.shape
```

```python
pca_values_df = pd.DataFrame(data = components, columns = ['Principal Component 1', 'Principal Component 2', 
                                                           'Principal Component 3', 'Principal Component 4'])
pca_values_df
```

### Algorithm #2: Partial Least Squares (PLS)

```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict
```

```python
# The goal is to predict y from x
X = np.array([pca_df['glas'], pca_df['crtp'], pca_df['mica'], pca_df['amtp']]) # x matrix, predictors var
Y = np.array([pca_df['wrp1'], pca_df['wrp2'], pca_df['wrp3'], pca_df['wrp4'], pca_df['wrp5'], pca_df['wrp6'], 
     pca_df['wrp7'], pca_df['wrp8'], 
     pca_df['st1'], pca_df['st2'], pca_df['st3'], pca_df['st4'], pca_df['st5'], pca_df['st6']]) # y matrix, reponse var 
```

```python
# Define PLS object and choose number of components
pls = PLSRegression(n_components=4)
```

```python
X.transpose().shape
```

```python
X.transpose()
```

```python
Y.transpose().shape
```

```python
Y.transpose()
```

```python
# Fit
pls.fit(X.transpose(), Y.transpose())
```

```python
# Cross-validation
y_cv = cross_val_predict(pls, X.transpose(), Y.transpose(), cv=10)
y_cv
```

```python
# Calculate r^2 score, goodness of fit
score = r2_score(Y.transpose(), y_cv)
score
```

```python
# mean squares error, average squared difference between the estimated values and the actual value
mse = mean_squared_error(Y.transpose(), y_cv)
mse
```
